## Regression

```python
import time
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
from sklearn import preprocessing

def SimuLinear(N,M,N1,SNR, seed=1):
    np.random.seed(seed)
    X =  np.random.normal(0, 1, (N, M))
    np.random.rand(seed*2+1)
    X1 =  np.random.normal(0, 1, (N1, M))
    M1=int(0.05*M)
    beta = np.append(np.append(np.array([SNR]),np.random.normal(5, 1,M1-1)),np.array([0]*(M-M1)))

    Y = np.dot(X,beta)+np.random.normal(0, 1,N)
    Y1 = np.dot(X1,beta)+np.random.normal(0, 1,N1)

    Y = preprocessing.scale(Y)
    Y1 = preprocessing.scale(Y1)

    return [X,Y,X1,Y1]


N=500 
M=200
B=100

n_ratio=int(np.sqrt(N))/N
m_ratio=int(np.sqrt(M))/M

snr = 5
itrial=1

fit_func = DecisionTreeReg

print('LOCOMP')
X,Y,X1,Y1 =SimuLinear(N,M,B,snr,seed = 123*itrial+456)
```

```python
x=LOCOMPReg(X,Y,n_ratio,m_ratio,B,fit_func , selected_features=[],alpha=0.1,bonf=True)
x.run_loco()
x.loco_ci
```

## correct variance

```python
x.correct_variance(eps=0.00005)
```

## Classification

```python
import time
import pandas as pd
import numpy as np

def SimuLinearClass(N,M,N1,SNR,M1=5,seed=1):
    np.random.seed(seed)
    X =  np.random.normal(0, 1, (N, M))
    np.random.rand(seed*2+1)
    X1 =  np.random.normal(0, 1, (N1, M))
    M1=int(0.05*M)
    beta = np.append(np.append(np.array([SNR]),np.random.normal(5, 1,M1-1)),np.array([0]*(M-M1)))
    pr = np.dot(X,beta)+np.random.normal(0, 1,N)
    pr=(np.exp(pr)/(1+np.exp(pr)))
    Y = np.random.binomial(1,pr)

    pr1 = np.dot(X1,beta)+np.random.normal(0, 1,N1)
    pr1= (np.exp(pr1))/(1+np.exp(pr1))
    Y1 = np.random.binomial(1,pr1)
    return (X,Y,X1,Y1)

import warnings
warnings.filterwarnings("ignore")
N=500 
M=200
B=100

n_ratio=int(np.sqrt(N))/N
m_ratio=int(np.sqrt(M))/M

snr = 5
itrial=1

fit_func = DecisionTreeClass

print('LOCOMP')
X,Y,X1,Y1 =SimuLinearClass(N,M,B,snr,seed = 123*itrial+456)
```

```python
x=LOCOMPClass(X,Y,n_ratio,m_ratio,B,fit_func , selected_features=[],alpha=0.1,bonf=True)
x.run_loco()
x.loco_ci
```

```python
x.correct_variance(eps=0.00005)
```
